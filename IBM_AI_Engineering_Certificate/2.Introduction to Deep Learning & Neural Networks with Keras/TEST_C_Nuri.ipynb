```python
import pandas as pd
import numpy as np
import sklearn 
```

```python
concrete= pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete.head()
```

```python
concrete_columns = concrete.columns
X = concrete[concrete_columns[concrete_columns != 'Strength']] 
y = concrete['Strength']  
```

```python
X.head()
```

```python
y.head()
```

```python
X_norm = (X - X.mean()) / X.std()
X_norm.head()
```

```python
n_cols = X_norm.shape[1] # number of predictors
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=156)
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model_A():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = regression_model_A()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse_result = []
epochs = 50
for epochs in range (epochs):
    model.fit(X_train, y_train, epochs=100, verbose=2)
    pred = model.predict(X_test)
    mse = mean_squared_error(y_test, pred)
    mse_result.append(mse)

```

```python
print(mse_result)
```

```python
print('mean of mse:',np.mean(mse_result))
```
