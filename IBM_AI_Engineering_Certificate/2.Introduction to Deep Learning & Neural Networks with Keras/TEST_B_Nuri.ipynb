파트 A를 반복하지만 정규화된 버전의 데이터를 사용합니다. 
데이터를 정규화하는 한 가지 방법은 개별 예측 변수에서 평균을 뺀 다음 표준 편차로 나누는 것임을 기억하십시오.

평균 제곱 오차의 평균은 A 단계의 평균과 어떻게 비교됩니까?

```python
import pandas as pd
import numpy as np
import sklearn 
```

```python
concrete= pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete.head()
```

```python
concrete_columns = concrete.columns
X = concrete[concrete_columns[concrete_columns != 'Strength']] 
y = concrete['Strength']  
```

```python
X.head()
```

```python
y.head()
```

```python
X_norm = (X - X.mean()) / X.std()
X_norm.head()
```

```python
n_cols = X_norm.shape[1] # number of predictors
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=156)
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model_A():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = regression_model_A()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse_result = []
epochs = 50
for epochs in range (epochs):
    model.fit(X_train, y_train, epochs=50, verbose=2)
    pred = model.predict(X_test)
    mse = mean_squared_error(y_test, pred)
    mse_result.append(mse)

```

```python
print(mse_result)
```

```python
print('mean of mse:',np.mean(mse_result))
```

```python
# Compared to A(mean of mse) with B(mean of mse(normalized)),
# B(normalized) is bigger than A
```
